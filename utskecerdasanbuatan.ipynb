Aturan Logika
Berikut adalah empat aturan berdasarkan gejala yang mungkin muncul:

1.   Aturan 1: Jika P1 (daun menguning) dan P2 (terdapat bercak hitam), maka hama yang mungkin adalah Hama Jamur.
2.   Aturan 2: Jika P3 (daun berlubang) dan P4 (tanaman layu), maka hama yang mungkin adalah Ulat.
3.  Aturan 3: Jika P1 (daun menguning) dan P4 (tanaman layu), maka hama yang mungkin adalah Hama Kutu.
4.  Aturan 4: Jika P2 (terdapat bercak hitam) dan P3 (daun berlubang), maka hama yang mungkin adalah Hama Kutu Daun.


Sistem pakar ini menggunakan logika proposisional untuk mengidentifikasi jenis hama pada tanaman berdasarkan gejala yang diamati. Dalam notasi proposisional, kita mendefinisikan beberapa proposisi:
ğ‘ƒ1untuk daun menguning,
ğ‘ƒ2 untuk bercak hitam,
ğ‘ƒ3 untuk daun berlubang, dan
ğ‘ƒ4 untuk tanaman layu.
Terdapat beberapa aturan yang ditetapkan, seperti aturan pertama yang menyatakan bahwa jika
ğ‘ƒ1 dan ğ‘ƒ2 benar, maka hama yang mungkin adalah "Hama Jamur". Aturan kedua menunjukkan bahwa jika
ğ‘ƒ3 dan ğ‘ƒ4 benar, maka output adalah "Ulat". Aturan ketiga menyatakan bahwa jika ğ‘ƒ1dan ğ‘ƒ4 benar, hama yang teridentifikasi adalah "Hama Kutu", sedangkan aturan keempat menyatakan bahwa jika ğ‘ƒ2 dan ğ‘ƒ3 benar, maka hama tersebut adalah "Hama Kutu Daun". Jika tidak ada aturan yang terpenuhi, sistem akan menginformasikan bahwa hama tidak dikenali. Dengan demikian, sistem ini memeriksa kombinasi proposisi berdasarkan nilai yang diberikan, dan hasilnya memberikan petani pemahaman yang jelas tentang jenis hama yang mungkin menyerang tanaman mereka.

```python
def diagnose_pest(symptoms):
    # Aturan 1
    if "daun menguning" in symptoms and "bercak hitam" in symptoms:
        return "Kemungkinan hama: Hama Jamur"

    # Aturan 2
    elif "daun berlubang" in symptoms:
        return "Kemungkinan hama: Hama Ulat"

    # Aturan 3
    elif "daun menguning" in symptoms and "tanaman layu" in symptoms:
        return "Kemungkinan hama: Hama Kutu"

    # Aturan 4
    elif "bercak hitam" in symptoms and "daun berlubang" in symptoms:
        return "Kemungkinan hama: Hama Kutu Daun"

    # Jika tidak ada kecocokan
    else:
        return "Hama tidak teridentifikasi"

# Contoh penggunaan
gejala = ["daun menguning", "tanaman layu"]  # Menambahkan "tanaman layu"
hasil = diagnose_pest(gejala)
print(hasil)
```

```python
def identifikasi_hama(daun_menguning, bercak_hitam, daun_berlubang, tanaman_layu):
    # Aturan 1: Hama Jamur
    if daun_menguning and bercak_hitam:
        return "Hama Jamur"
    # Aturan 2: Ulat
    elif daun_berlubang and tanaman_layu:
        return "Ulat"
    # Aturan 3: Hama Kutu
    elif daun_menguning and tanaman_layu:
        return "Hama Kutu"
    # Aturan 4: Hama Kutu Daun
    elif bercak_hitam and daun_berlubang:
        return "Hama Kutu Daun"
    else:
        return "Hama tidak dikenali"

# Contoh penggunaan
gejala = {
    "daun_menguning": True,
    "bercak_hitam": False,
    "daun_berlubang": False,
    "tanaman_layu": True
}

hasil = identifikasi_hama(
    gejala["daun_menguning"],
    gejala["bercak_hitam"],
    gejala["daun_berlubang"],
    gejala["tanaman_layu"]
)

print("Hasil Diagnosa:", hasil)

# Tabel hasil diagnosa
tabel_hama = [
    ["No", "Gejala", "Hama yang Mungkin"],
    [1, "Daun Menguning + Bercak Hitam", "Hama Jamur"],
    [2, "Daun Berlubang + Tanaman Layu", "Ulat"],
    [3, "Daun Menguning + Tanaman Layu", "Hama Kutu"],
    [4, "Bercak Hitam + Daun Berlubang", "Hama Kutu Daun"],
    [5, "Tidak ada gejala yang dikenali", "Hama tidak dikenali"]
]

# Mencetak tabel
print("\nTabel Diagnosa Hama:")
for row in tabel_hama:
    print("{:<3} {:<40} {:<20}".format(row[0], row[1], row[2]))
```

Sistem pakar ini dirancang untuk membantu petani mengidentifikasi jenis hama pada tanaman berdasarkan gejala yang diamati. Petani memberikan gejala dalam bentuk boolean, seperti apakah daun menguning, terdapat bercak hitam, daun berlubang, atau tanaman layu. Misalnya, jika daun_menguning adalah True dan tanaman_layu adalah True, maka fungsi identifikasi_hama akan dipanggil dengan parameter tersebut. Fungsi ini kemudian mengevaluasi setiap aturan yang telah ditetapkan secara berurutan. Aturan pertama mengindikasikan bahwa jika daun menguning dan terdapat bercak hitam, output yang dihasilkan adalah "Hama Jamur". Jika kombinasi gejala lain terpenuhi, seperti daun berlubang dan tanaman layu, outputnya adalah "Ulat". Jika tidak ada aturan yang cocok, sistem akan menginformasikan bahwa hama tidak dikenali. Setelah memeriksa semua aturan, sistem memberikan hasil diagnosa yang sesuai. Selain itu, sistem juga menyajikan tabel yang merangkum semua kombinasi gejala dan jenis hama yang mungkin, memberikan pemahaman yang lebih baik kepada petani tentang masalah yang dihadapi tanaman mereka. Dengan demikian, sistem ini menyediakan solusi yang efektif dan efisien dalam mengatasi hama tanaman, memungkinkan petani untuk membuat keputusan yang lebih baik dalam perawatan tanaman mereka.
